# _Loading the necessary libraries and the dataset._

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the dataset
data = pd.read_csv('data/housing.csv')
```

# _Preprocessing the data._

```python
X_income = data[['median_income']]
X_population = data[['population']]
X_households = data[['households']]
y = data['median_house_value']

# Check for missing values
print(data.isnull().sum())

# Check the data types and statistics
print(data.dtypes)
print(data.describe())
```

# _Converting data into training and test sets._

```python
X_income_train, X_income_test, y_train, y_test = train_test_split(X_income, y, test_size=0.2, random_state=42)
X_population_train, X_population_test, y_train, y_test = train_test_split(X_population, y, test_size=0.2, random_state=42)
X_households_train, X_households_test, y_train, y_test = train_test_split(X_households, y, test_size=0.2, random_state=42)
```

# _Creating a linear regression model for each independent variable_

```python
model_income = LinearRegression()
model_population = LinearRegression()
model_households = LinearRegression()
```

# _Training the models on the training data_

```python
model_income.fit(X_income_train, y_train)
model_population.fit(X_population_train, y_train)
model_households.fit(X_households_train, y_train)
```

# _Evaluating the models on the test data and interpret the results._

```python
# Evaluate the models on the test data
income_predictions = model_income.predict(X_income_test)
population_predictions = model_population.predict(X_population_test)
households_predictions = model_households.predict(X_households_test)

# Plot the actual values vs predicted values
plt.scatter(y_test, income_predictions)
plt.xlabel('Actual Median House Value')
plt.ylabel('Predicted Median House Value')
plt.title('Linear Regression - Median Income')
plt.show()

plt.scatter(y_test, population_predictions)
plt.xlabel('Actual Median House Value')
plt.ylabel('Predicted Median House Value')
plt.title('Linear Regression - Population')
plt.show()

plt.scatter(y_test, households_predictions)
plt.xlabel('Actual Median House Value')
plt.ylabel('Predicted Median House Value')
plt.title('Linear Regression - Number of Households')
plt.show()
```
